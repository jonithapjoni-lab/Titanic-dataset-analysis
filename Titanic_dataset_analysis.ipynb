```python
# 1. DATA IMPORT
import pandas as pd
df = pd.read_csv("/content/sample_data/mnist_train_small.csv")
print("Dataset Loaded Successfully! Displaying the first 5 rows:")
display(df.head())
```

```python
# 2. SERIES AND DATAFRAME CREATION
# DataFrame already created as df

# Assuming the first column '6' is the label, and '0' and '0.1' are pixel values
label_series = df[df.columns[0]] # Get the first column as label
pixel_series_1 = df[df.columns[1]] # Get the second column as a pixel feature
pixel_series_2 = df[df.columns[2]] # Get the third column as another pixel feature

print("Label Series (first column):", label_series.head())
print("\nPixel Series 1 (second column):", pixel_series_1.head())
print("\nPixel Series 2 (third column):", pixel_series_2.head())
```

```python
# 3. DATA INSPECTION

print("First 5 Rows:\n")
print(df.head())

print("\nDataset Info:\n")
print(df.info())

print("\nStatistical Summary:\n")
print(df.describe())
```

```python
# 4. HANDLING MISSING DATA

# Check missing values
print("\nMissing Values:\n")
print(df.isnull().sum())

# As there are no missing values in this dataset, no imputation or dropping is required.
# If this were the Titanic dataset, the following lines would be relevant:
# df["Age"].fillna(df["Age"].mean(), inplace=True)
# df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
# df.dropna(subset=["Fare"], inplace=True)

print("\nMissing Values After Cleaning (if any):")
print(df.isnull().sum())
```

```python
# 5. INDEXING AND SLICING

# Using loc (label-based)
print("\nRow with index 5 using loc:\n")
print(df.loc[5])

# Using iloc (position-based)
print("\nRow at position 5 using iloc:\n")
print(df.iloc[5])

# Extract specific columns using their actual names or positions
# Assuming the first column is the label and the next two are pixel values
print("\nFirst three columns (label and first two pixel features):\n")
print(df[[df.columns[0], df.columns[1], df.columns[2]]].head())
```

```python
# Rename the first column to 'label' as it represents the digit label
df.rename(columns={df.columns[0]: "label"}, inplace=True)

print("\nColumns after Renaming:\n")
print(df.columns)

# Create a pivot table to count the occurrences of each digit label
# This shows the distribution of digits in the dataset
pivot_table = df.pivot_table(values=df.columns[1], index="label", aggfunc="count") # Using the second column (first pixel value) for counting
print("\nPivot Table (Digit Counts):\n")
print(pivot_table)
```

```python
# 7. FILTERING AND SORTING

# Filter for images with label '5'
label_5_images = df[df["label"] == 5]

print("\nImages with label '5' (first 5 rows):\n")
display(label_5_images.head())

# Sorting by 'label' in ascending order
sorted_by_label = df.sort_values(by="label", ascending=True)
print("\nDataFrame sorted by 'label' (first 5 rows):\n")
display(sorted_by_label.head())
```

```python
# 8. GROUPING AND AGGREGATION

# Group by 'label' and calculate the mean of a pixel value
# Using the first pixel column (df.columns[1]) as an example
mean_pixel_by_label = df.groupby("label")[df.columns[1]].mean()
print("\nMean of first pixel value by label:\n", mean_pixel_by_label)

# Group by 'label' and count occurrences (similar to pivot table but using groupby)
label_counts = df.groupby("label").size()
print("\nCounts of each label:\n", label_counts)
```

```python
# Concat (example splitting and combining)
df_part1 = df.iloc[:200]
df_part2 = df.iloc[200:]
concat_df = pd.concat([df_part1, df_part2])

print("\nConcatenated DataFrame Shape:", concat_df.shape)
```

```python
# 10. CALCULATIONS AND INSIGHTS

# Most frequent digit
most_frequent_digit = df["label"].mode()[0]

# Least frequent digit
# Using value_counts() and taking the last element after sorting
least_frequent_digit = df["label"].value_counts(ascending=True).index[0]

# Average value of the first pixel column (example for a pixel feature)
average_pixel_value = df[df.columns[1]].mean()

# Total number of images
total_images = len(df)

print("\n===== INSIGHTS ====")
print(f"Most Frequent Digit: {most_frequent_digit}")
print(f"Least Frequent Digit: {least_frequent_digit}")
print(f"Average Value of First Pixel Column: {average_pixel_value:.2f}")
print(f"Total Number of Images: {total_images}")
```
